     - 서로 속도가 다르니 최적의 솔루션에 도달하는 데에 좋은 영향을 미침

### 단일층 클래스

````python
class SingleLayer:
    def __init__(self, learning_rate=0.1, l1=0, l2=0):
        self.w = None
        self.b = None
        self.losses = []
        self.val_losses = []
        self.w_history = []
        self.lr = learning_rate
        self.l1 = l1
        self.l2 = l2
        
    def forpass(self, x):
        z = np.sum(x * self.w) + self.b  # 직선 방정식 계산
        return z
    
    def backprop(self, x, err):
        w_grad = x * err  # 가중치에 대한 그래디언트 계산
        b_grad = 1 * err  # 절편에 대한 그래디언트를 계산
        return w_grad, b_grad
    
    def activation(self, z):
        z = np.clip(z, -100, None)  # 안전한 .np.exp() 하려고
        a = 1 / (1 + np.exp(-z))  # 시그모이드 게산
        return a
    
    def fit(self, x, y, epochs=100, x_val=None, y_val=None):
        self.w = np.ones(x.shape[1])  # 가중치 초기화
        self.b = 0  # 절편을 초기화
        self.w_history.append(self.w.copy())  # 가중치 기록
        np.random.seed(42)  # 랜덤 시드 지정
        for i in range(epochs):
            loss = 0
            # 인덱스 쉑쉑
            indexes = np.random.permutation(np.arange(len(x)))
            for i in indexes:  # 샘플에 대해 반복
                z = self.forpass(x[i])  # 정방향 계산
                a = self.activation(z)  # 활성화 함수 적용
                err = -(y[i] - a)  # 오차 계산
                w_grad, b_grad = self.backprop(x[i], err)  # 역방향 계산
                # 그레디언트에서 페널티 향의 미분 값 더함
                w_grad += self.l1 * np.sign(self.w) + self.l2 * self.w
                self.w -= self.lr * w_grad  # 가중치 업데이트
                self.b -= b_grad  # 절편 업데이트
                # 가중치를 기록
                self.w_history.append(self.w.copy())
                # 안전한 로그 계산을 위해 클리핑한 후 손실 누적
                a = np.clip(a, 1e-10, 1-1e-10)
                loss += -(y[i]*np.log(a)+(1-y[i])*np.log(1-a))
            # 에포크마다 평규 손실 저장
            self.losses.append(loss/len(y) + self.reg_loss())
            # 검증 세트에 대한 손실 계산
            self.update_val_loss(x_val, y_val)
            
    def predict(self, x):
        z = [self.forpass(x_i) for x_i in x]  # 정방향 계산
        return np.array(z) >= 0  # 스텝 함수
    
    def score(self, x, y):
        return np.mean(self.predict(x) == y)
    
    def reg_loss(self):
        return self.l1 * np.sum(np.abs(self.w)) + self.l2 / 2 * np.sum(self.w**2)
    
    def update_val_loss(self, x_val, y_val):
        if x_val is None:
            return
        val_loss = 0
        for i in range(len(x_val)):
            z = self.forpass(x_val[i])
            a = self.activation(z)
            a = np.clip(a, 1e-10, 1-1e-10)
            val_loss += -(y_val[i]*np.log(a)+(1-y_val[i])*np.log(1-a))
        self.val_losses.append(val_loss/len(y_val) + self.reg_loss())
````

````python
layer1 = SingleLayer()
layer1.fit(x_train, y_train)
layer1.score(x_val, y_val)
````

- 적절한 학습률로 가중치의 변화를 안정적으로 만들어서 전역 최솟값 도달

## 5-2 과대적합과 과소적합
- 과대적합: 훈련세트 정확도와 검증세트 정확도에 다소 차이가 있음(분산이 큼)
- 과소적합: 훈련세트 정확도와 검증세트 정확도의 분산이 작지만, 기대 정확도 보다 예상 정확도가 낮을 때(편향이 큼)

- 3번째 그래프가 아주 굿~


일반적인 방법: 모델 과대적합(뉴런,층++) -> 규제(가중치규제,드롭아웃) -> 적절한 모델

### 에포크 vs 손실함수
- 최적점 기준 많이 훈련 -> 과대적합
- 최적점 기준 적게 훈련 -> 과소적합
- 에포크를 모델 복잡도로 간주 가능

## 5-3 규제를 단일층 신경망에 적용
- 그래프 기울기가 작을수록(가중치down) 일반화 좋음
- 규제가 정확도에 영향 X

### L1 규제(손실함수 + L1노름)
- 일반적인 최적화 규제
- 손실함수에 L2노름 절대값 추가 -> alpha에 가중치 부호 곱하여 그레디언트 update
- 알파값(하이퍼파라미터)으로 가중치 규제
- 가중치가 커질수록 손실함수가 커지니까 그거 막으려고 규제
- 영향력 적음

``` w_grad += alpha * np.sign(w)```

선형회귀 + L1 = 라쏘(Lasso)

### L2 규제(손실함수 + L2노름)
- 손실함수에 L2노름에 제곱 추가 -> 1/2alpha에 가중치 곱하여 그레디언트 update
- 규제효과가 큼
- 규제가 많이 되어도 가중치를 절대 0으로 만들지 않아 L1보다 많이 선호

```w_grad += alpha * w```

선형회귀 + L2규제 = 릿지(Ridge)
선형휘귀 + L1규제 + L2규제 = 엘라스틱넷(ElasticNet)

````python
l1_list = [0.0001, 0.001, 0.01]

for l1 in l1_list:
    lyr = SingleLayer(l1=l1)
    lyr.fit(x_train_scaled, y_train, x_val=x_val_scaled, y_val=y_val)
    
    plt.plot(lyr.losses)
    plt.plot(lyr.val_losses)
    plt.title('Learning Curve (l1={})'.format(l1))
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train_loss', 'val_loss'])
    plt.ylim(0, 0.3)
    plt.show()
    
    plt.plot(lyr.w, 'bo')
    plt.title('Weight (l1={})'.format(l1))
    plt.ylabel('value')
    plt.xlabel('weight')
    plt.ylim(-4, 4)
    plt.show()
````

````python
layer5 = SingleLayer(l1=0.001)
layer5.fit(x_train_scaled, y_train, epochs=20)
layer5.score(x_val_scaled, y_val)
````

````python
l2_list = [0.0001, 0.001, 0.01]

for l2 in l2_list:
    lyr = SingleLayer(l2=l2)
    lyr.fit(x_train_scaled, y_train, x_val=x_val_scaled, y_val=y_val)
    
    plt.plot(lyr.losses)
    plt.plot(lyr.val_losses)
    plt.title('Learning Curve (l2={})'.format(l2))
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train_loss', 'val_loss'])
    plt.ylim(0, 0.3)
    plt.show()
    
    plt.plot(lyr.w, 'bo')
    plt.title('Weight (l2={})'.format(l2))
    plt.ylabel('value')
    plt.xlabel('weight')
    plt.ylim(-4, 4)
    plt.show()
````

````python
layer6 = SingleLayer(l2=0.01)
layer6.fit(x_train_scaled, y_train, epochs=50)
layer6.score(x_val_scaled, y_val)
````

## 5-4 교차 검증
### 과정
- 훈련 세트를 k개의 fold 나누기
- 첫번째 폴드를 검증 세트로 사용하고 나머지 폴드를 훈련 세트 사용
- 모델을 훈련한 다음에 검증 세트로 평가
- 차례대로 다음 폴드를 검증 세트 사용하여 반복
- k개의 검증 세트로 k번 성능을 평가한 후 계산된 성능의 평균을 내어 최종 성능을 계산
- 딥러닝은 아니지만, 머신러닝에서는 필수

````python
# 10개의 폴드로 사이킷런 교차 검증 수행
from sklearn.model_selection import cross_validate
sgd = SGDClassifier(loss='log', penalty='l2', alpha=0.001, random_state=42)
scores = cross_validate(sgd, x_train_all, y_train_all, cv=10)
print(np.mean(scores['test_score']))
````

### 전처리 단계를 포함해 교차 검증 수행
- 검증 세트를 따로 나누지 않고 훈련 세트 전체 사용
- 전처리와 모델을 하나의 파이프라인으로 정의
- 이유는 아까 표준화 설명 참고 (표준화 평균, 표준편차 통일)

````python
# 파이프라인 안에서 표준화
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
pipe = make_pipeline(StandardScaler(), sgd)
scores = cross_validate(pipe, x_train_all, y_train_all, cv=10, return_train_score=True)
print(np.mean(scores['test_score']))
````
