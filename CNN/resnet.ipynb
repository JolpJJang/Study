```
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Activation, ZeroPadding2D, Conv2D, Flatten, Dense, MaxPooling2D, GlobalAveragePooling2D, BatchNormalization, Add
from tensorflow.keras import Sequential
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dropout
from keras import Input
```

```
def conv1(model):
  model.add(Conv2D(64, (7, 7), strides=2, padding='valid', input_shape=(230,230,3)))
  model.add(BatchNormalization())
  model.add(Activation('relu'))
  model.add(ZeroPadding2D(padding = 1))
  
  return model

def conv2_x(model):
  model.add(MaxPooling2D((3, 3), strides=2))
  shortcut = model
  for i in range(3):
    if (i==0):
      model.add(Conv2D(64, (1,  1), strides=1))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(64, (3, 3), strides=1, padding='same'))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(256, (1, 1), strides=1))
      model.add(BatchNormalization())

      shortcut = Conv2D(256, (1, 1), strides=1)(shortcut)
      shortcut = BatchNormalization()(shortcut)
      
      model.Add([model, shortcut])
      
      model.add(Activation('relu'))

      shortcut = model

    else:
      model.add(Conv2D(64, (1,  1), strides=1))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(64, (3, 3), strides=1, padding='same'))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(256, (1, 1), strides=1))
      model.add(BatchNormalization())

      model.Add([model, shortcut])

      model.add(Activation('relu'))

      shortcut = model

  return model

def conv3_x(model):
  shortcut = model
  for i in range(8):
    if (i==0):
      model.add(Conv2D(128, (1,  1), strides=2))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(128, (3, 3), strides=1, padding='same'))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(512, (1, 1), strides=1))
      model.add(BatchNormalization())

      shortcut = Conv2D(512, (1, 1), strides=2)(shortcut)
      shortcut = BatchNormalization()(shortcut)
      
      model.Add([model, shortcut])
      
      model.add(Activation('relu'))

      shortcut = model

    else:
      model.add(Conv2D(128, (1,  1), strides=1))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(128, (3, 3), strides=1, padding='same'))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(512, (1, 1), strides=1))
      model.add(BatchNormalization())

      model.Add([model, shortcut])

      model.add(Activation('relu'))

      shortcut = model

  return model

def conv4_x(model):
  shortcut = model
  for i in range(36):
    if (i==0):
      model.add(Conv2D(256, (1,  1), strides=2))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(256, (3, 3), strides=1, padding='same'))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(1024, (1, 1), strides=1))
      model.add(BatchNormalization())

      shortcut = Conv2D(1024, (1, 1), strides=2)(shortcut)
      shortcut = BatchNormalization()(shortcut)
      
      model.Add([model, shortcut])
      
      model.add(Activation('relu'))

      shortcut = model

    else:
      model.add(Conv2D(256, (1,  1), strides=1))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(256, (3, 3), strides=1, padding='same'))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(1024, (1, 1), strides=1))
      model.add(BatchNormalization())

      model.Add([model, shortcut])

      model.add(Activation('relu'))

      shortcut = model

  return model

def conv5_x(model):
  shortcut = model
  for i in range(3):
    if (i==0):
      model.add(Conv2D(512, (1,  1), strides=2))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(512, (3, 3), strides=1, padding='same'))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(2048, (1, 1), strides=1))
      model.add(BatchNormalization())

      shortcut = Conv2D(2048, (1, 1), strides=2)(shortcut)
      shortcut = BatchNormalization()(shortcut)
      
      model.Add([model, shortcut])
      
      model.add(Activation('relu'))

      shortcut = model

    else:
      model.add(Conv2D(512, (1,  1), strides=1))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(512, (3, 3), strides=1, padding='same'))
      model.add(BatchNormalization())
      model.add(Activation('relu'))

      model.add(Conv2D(2048, (1, 1), strides=1))
      model.add(BatchNormalization())

      model.Add([model, shortcut])

      model.add(Activation('relu'))

      shortcut = model
    
  return model


def ResNet():
  #input = Input(shape=(224, 224, 3), dtype='float32', name='input')
  #input = ZeroPadding2D(padding=(3, 3))(input)

  res = tf.keras.Sequential()
  res = conv1(res)
  res = conv2_x(res)
  res = conv3_x(res)
  res = conv4_x(res)
  res = conv5_x(res)

  res.add(AveragePooling2D())
  res.add(Flatten())
  res.add(Dense(1000, activation='softmax'))

  return res
```
